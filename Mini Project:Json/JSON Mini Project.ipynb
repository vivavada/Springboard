```python
# load json as string
file=json.load((open('/Users/inan/Desktop/data_wrangling_json/data/world_bank_projects.json')))
```

```python
j_s=json_normalize(file,'mjtheme_namecode')
```

```python
j_s
```

```python
# I dropped the empty rows and duplicates to create clean dictionary to be able to use missing values later.

jsn=j_s[j_s.name!='']
```

```python
js=jsn.drop_duplicates()
js
```

```python
# I will use the dictionary to fill the missing data using for loop which i created below
dic=js.set_index('code').to_dict()['name']
```

```python
dic
```

```python
#3-You will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.
for key, value in dic.items():
    j_s.loc[j_s.code == key,'name'] = value
j_s
```
